# Producer_3

        This producer loads the hotspot_TERRA_streaming.csv and randomly (with replacement) feed the data to
        the stream every 2 seconds. Additional information, such as Sender_id and Created_time, will be appended.

```python
import csv
import datetime as dt
from time import sleep
from json import dumps
from kafka import KafkaProducer
import random


data=[]
with open('hotspot_TERRA_streaming.csv') as f:
    reader = csv.reader(f)
    next(reader, None) # avoid the header
    
    for row in reader: 
        d={'Latitude':float(row[0]),
           'Longitude':float(row[1]),
           'Confidence':int(row[2]),
           'Surface_Temperature_Celcius':int(row[3]),
           'Sender_ID':3     
            }
        data.append(d)



def publish_message(producer_instance, topic_name, data):
    try:
        value_bytes = bytes(data, encoding='utf-8')
        producer_instance.send(topic_name, value=value_bytes)
        print('Message published successfully. ' + data)
    except Exception as ex:
        print('Exception in publishing message.')
        print(str(ex))
        
def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=['127.0.0.1:9092'],
                                  api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka.')
        print(str(ex))
    finally:
        return _producer
    
if __name__ == '__main__':
   
    topic = 'vic_fire_producer'
    print('Publishing records..')
    producer03 = connect_kafka_producer()

    
    for item in random.sample(data,1800): #streaming for 6 m
        item['Created_Date']=dt.datetime.now().strftime("%d/%m/%Y")
        item['Created_Time']=dt.datetime.now().strftime("%X")
        send_data = dumps(item)
        
        publish_message(producer03, topic, send_data) 
        sleep(2) #send every 2 seconds
```
